```python
# -----------------------------
# Python お試しコード
# -----------------------------

# 関数定義
def greet(name):
    return f"こんにちは、{name}さん！"

# 計算関数
def calc(x, y):
    return x * y + 3

# メインメソッド
def main():
    print("== お試しコード実行中 ==")
    print(greet("金本"))
    print("計算結果:", calc(4, 7))
    print("完了！")

# メインブロック
if __name__ == "__main__":
    main()
```
